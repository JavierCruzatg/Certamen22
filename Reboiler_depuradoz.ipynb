```python
df1 = df1
df1.set_index([
	'Fecha',
], inplace=True)
attrs = {
	'var': 'df1',
	'type': type(df1).__name__,
}
[attrs]
df1
```

```python
cond = ((df1['Conductividad'].le(12)))
df1 = df1[cond]
attrs = {
	'var': 'df1',
	'type': type(df1).__name__,
}
[attrs]
df1
```

```python
from functools import partial
from pandas.tseries.frequencies import to_offset

def roundfunc(t, freq):
	freq = to_offset(freq)
	return pd.Timestamp((t.value // freq.delta.value) * freq.delta.value)

df1 = df1.groupby(partial(roundfunc, freq='120T'))

attrs = {
	'var': 'df1',
	'type': type(df1).__name__,
}
[attrs]
df1
enum = enumerate(df1.groups)
nextel = next(enum)
nextel[1]
df1.get_group(nextel[1])
nextel = next(enum)
nextel[1]
df1.get_group(nextel[1])
```

```python
df1 = df1.aggregate('median')

attrs = {
	'var': 'df1',
	'type': type(df1).__name__,
}
[attrs]
df1
```
