```python
import os
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler
import joblib
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
file = os.path.join('C:/Users/javic/Desktop/notebook para repositorio/Chancador', 'Chancadorr')
df15 = pd.read_csv(file)
attrs = {
	'var': 'df15',
	'type': type(df15).__name__,
}
[attrs]
df15
```

```python
df15 = df15
s = df15['Tiempo']
df15['Tiempo'] = pd.to_datetime(s, errors='coerce')

attrs = {
	'var': 'df15',
	'type': type(df15).__name__,
}
[attrs]
df15
```

```python
df15 = df15
df15.set_index([
	'Tiempo',
], inplace=True)
attrs = {
	'var': 'df15',
	'type': type(df15).__name__,
}
[attrs]
df15
```

```python
col_start = df9['Inicio']
col_end = df9['Fin']
idx = pd.IntervalIndex.from_arrays(col_start, col_end, closed='left')
df9_interval = df9.set_index(idx)
df9_interval = df9_interval.drop(columns=[
	'Inicio',
	'Fin'
], axis='columns')
df_labels14 = pd.DataFrame(index=df15.index)
df_labels14['Estado de salud'] = 'Operativa'
for i, r in df9_interval.iterrows():
	mask = (df_labels14.index >= i.left) & (df_labels14.index < i.right)
	df_labels14.loc[mask, 'Estado de salud'] = r['Tipo de Detención']
attrs = {
	'var': 'df_labels14',
	'type': type(df_labels14).__name__,
}
[attrs]
df_labels14
```

```python
df18 = pd.concat([
	df15,
	df_labels14,
], axis='columns', join='outer')

attrs = {
	'var': 'df18',
	'type': type(df18).__name__,
}
[attrs]
df18
```

```python
cond = ((df18['Estado de salud'].ne('Pausa')))
df18 = df18[cond]
attrs = {
	'var': 'df18',
	'type': type(df18).__name__,
}
[attrs]
df18
```

```python
df18 = df18
cond = ((df18['Estado de salud'].eq('Operativo')))
df18['Estado de salud'].mask(cond, other=0, inplace=True)

cond = ((df18['Estado de salud'].eq('falla')))
df18['Estado de salud'].mask(cond, other=1, inplace=True)

attrs = {
	'var': 'df18',
	'type': type(df18).__name__,
}
[attrs]
df18
```

```python
file = os.path.join('C:/Users/javic/Desktop/notebook para repositorio/Chancador', 'Chancador depurado')
df18.to_csv(file)
attrs = {
	'var': 'empty',
	'type': 'empty',
}
[attrs]
'csv file saved'
```
